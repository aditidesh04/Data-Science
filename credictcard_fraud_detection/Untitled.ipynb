Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X = new_data.drop(columns='Class' , axis = 1)
y=new_data['Class']
```

```python
print(X)
```

Split the data in testing data and training data



```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
print(X.shape, X_train.shape ,X_test.shape)
```

Model Training through LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model=LogisticRegression()
model.fit(X_train,y_train)
```

Model Evaluation - Accuracy Test

```python
from sklearn.metrics import accuracy_score
```

```python
X_train_prediction = model.predict(X_train)
X_train_accuracy = accuracy_score(X_train_prediction,y_train)
```

```python
print('Accuracy Score on training data : ',X_train_accuracy*100)
```

```python
X_test_prediction = model.predict(X_test)
X_test_accuracy = accuracy_score(X_test_prediction, y_test)
```

```python
print('Accuracy Score on testing data: ', X_test_accuracy*100)
```
